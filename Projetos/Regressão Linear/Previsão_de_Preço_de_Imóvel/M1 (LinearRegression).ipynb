### Limpeza e Tratamento da Base de Dados
##### Momento que limpamos as colunas que contem varios dados nulos, tratamos as colunas que contem dado faltantes e concertamos as anomalias para fazer o treinamento do modelo.

```python
# REMOÇÃO DAS COLUNAS QUE TEM MUITOS DADOS FALTANTES
dados = dados.drop(['Lot Frontage', 'Alley', 'Fireplace Qu', 'Pool QC', 'Fence', 'Misc Feature'], axis=1)
```

```python
# ANÁLISE DE DADOS NULOS
dados_nulos = pd.DataFrame(dados.isnull().sum(), columns=['Dados Nulos'])
dados_nulos = dados_nulos.loc[dados_nulos['Dados Nulos'] > 0]
```

```python
# TRATAMENTO DAS COLUNAS COM POUCOS DADOS NULOS
for i in dados_nulos.index:
    dados[i].fillna(sts.mode(dados[i]), inplace=True)
```

```python
# CHECANDO SE AINDA EXISTE DADOS NULOS NOVAMENTE
dados_nulos = pd.DataFrame(dados.isnull().sum(), columns=['Dados Nulos'])
dados_nulos.loc[dados_nulos['Dados Nulos'] > 0]
```

##### Todos os dados nulos foram removidos ou tratados da nossa base de dados.

___
### Criação do Modelo de Previsão
##### Agora vamos criar o modelo que vai prever os valores das casas de acordo com seus atributos.

```python
dados.shape[1]
```

```python
# SEPARAÇÃO DOS DADOS EM PREVISORES E CLASSES
previsores = dados.iloc[:, 0:dados.shape[1]-1].values
classe = dados.iloc[:, dados.shape[1]-1].values
```

```python
# TRANSFORMAÇÃO DOS ATRIBUTOS CATEGÓRICOS EM NUMÉRICOS
for i in range(previsores.shape[1]):
    if type((previsores[0:1, i])[0]) != int:
        labelencoder = LabelEncoder()
        previsores[:, i] = labelencoder.fit_transform(previsores[:, i])
```

```python
# SEPARAÇÃO DOS DADOS EM TREINO E TESTE (70%/30%)
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores,
                                                                  classe,
                                                                  test_size=.30)
```

```python
# CRIAÇÂO E TREINAMENTO DO MODELO
modelo = LinearRegression()
modelo.fit(X_treinamento, y_treinamento)
```

##### Com o modelo criado podemos agora fazer previsões, irei utilizar a primeira linha da tabela para fazer a previsão do seu valor e compará-la com o seu valor real.

```python
# PREVISÃO DO MODELO X VALOR REAL
previsao = modelo.predict([previsores[0]])[0]
valor_real = dados['SalePrice'][0]
print('PREVISÃO:', round(previsao,2), '\nVALOR REAL:', valor_real)
```

##### Podemos ver que o modelo chegou bem próximo do valor real da casa, vamos fazer um novo teste, utilizando outra casa da nossa base de dados e fazer a mesma comparação.

```python
# PREVISÃO DO MODELO X VALOR REAL
previsao = modelo.predict([previsores[25]])[0]
valor_real = dados['SalePrice'][25]
print('PREVISÃO:', round(previsao,2), '\nVALOR REAL:', valor_real)
```
